```python
age_group = Titanic_df["Age"]
def encode_age(age):
    if age < 18:
        return 'Child'
    elif age > 18 and age < 65:
        return 'Adult'
    else :
      return 'senior'

Titanic_df['Age_Group'] = Titanic_df['Age'].apply(encode_age)
Titanic_df[['Age', 'Age_Group']].head()
```
